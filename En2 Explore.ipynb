```python
import pandas as pd
import itertools
```

```python
df = pd.read_csv('en2.csv.gz', sep='\t')
df
```

```python
test = df['from'].str.split('_', expand=True).join(
    df['to'].str.split('_', expand=True), lsuffix='_l'
)

test.columns = range(len(test.columns))

```

```python
dfs = []

for i,j in itertools.combinations(range(len(test.columns)), 2):
    dfs.append(test[[i, j]].rename(columns={i: 'word1', j: 'word2'}).dropna().drop_duplicates())
    
expanded = pd.concat(dfs).drop_duplicates().reset_index(drop=True).dropna()

expanded
```

```python
expanded.to_csv('expanded_underscores.csv', index=False)
```

```python
dic = {}

#def add_to_dic(word1, word2):
#    if word1

#for i, row in expanded.iterrows():
    
for word, group in expanded.groupby('word1'):
    dic[word] = dict([(w[1]['word2'], 1) for w in group.iterrows()])
    
for word, group in expanded.groupby('word2'):
    dic[word] = dict([(w[1]['word1'], 1) for w in group.iterrows()])
```

```python
df = pd.concat([
    df.rename(columns={'to': 'word1', 'from': 'word2'}),
    df.rename(columns={'to': 'word2', 'from': 'word1'})
]).drop_duplicates().reset_index(drop=True).dropna()
```

```python
dic = {}
for word, group in df.groupby('word1'):
    dic[word] = dict([(w[1]['word2'], w[1]['relation']) for w in group.iterrows()])
```

```python
#board = ['time', 'france', 'mexico', 'date', 'yard', 'face', 'straw', 'plate', 'undertaker']
#board = ['web', 'ambulance', 'hospital', 'spell', 'lock', 'charge', 'tail', 'link', 'cook']
P1 = 1
P2 = -3
A = -10

'''board = {
    'web': P1,
    'ambulance': P1,
    'hospital': P1,
    'spell': P1,
    'lock': P1,
    'charge': P1,
    'tail': P1,
    'link': P1,
    'cook': P1,
    'time': P2,
    'france': P2,
    'mexico': P2,
    'date': P2,
    'yard': P2,
    'face': P2,
    'straw': P2,
    'plate': P2,
    'food': A
}'''

board = {
    'web': P1,
    'ambulance': P1,
    'hospital': P1,
    'spell': P1,
    'lock': P1,
    'charge': P1,
    'tail': P1,
    'link': P1,
    'cook': P1,
    'time': P2,
    'france': P2,
    'mexico': P2,
    'date': P2,
    'yard': P2,
    'face': P2,
    'straw': P2,
    'plate': P2,
    'food': A
}
results = {}

def add_word(w, value, ow):
    if w in results:
        #results[w] += [ow]
        results[w] += value
    else: 
        #results[w] = [ow]
        results[w] = value

for word in board:
    #for ass in df[df.word1 == word].word2.unique():
    for ass in dic[word]:
        add_word(ass, board[word], word)
        
def validate_word(word, board):
    for b in board:
        if word in b or b in word:
            return False
        #if 'FormOf' in list(df[(df['word1'] == word) & (df['word2'] == b)]['relation']):
        if (b in dic[word]) and (dic[word][b] == 'FormOf'):
            return False
        
    #print(word)
    return True

for word,sim in sorted(list(results.items()), key=lambda x: -x[1])[:20]:
    if validate_word(word, board):
        print(word,sim)
```
