```python
import pandas as pd
```

```python
df = pd.read_csv('conceptnet-assertions-5.5.0.csv.gz', sep='\t', header=None, names=['uri', 'relation', 'from', 'to', 'json'])

df.head()
```

```python
df = df[(df['from'].str.extract(r'/./([^/]*)/.*') == 'en') & (df['to'].str.extract(r'/./([^/]*)/.*') == 'en')]
```

```python
df = df[['relation', 'from', 'to']]
```

```python
df['from'] = df['from'].str.extract(r'/./[^/]*/([^/]*).*')
df['to'] = df['to'].str.extract(r'/./[^/]*/([^/]*).*')
```

```python
df.to_csv('en2.csv.gz', compression='gzip', index=False, encoding='utf-8', sep='\t')
```
