```python
import pandas as pd
```

```python
df = pd.read_csv('mappingbased_literals_en.ttl.bz2', sep=' ', compression='bz2', quotechar='"')
```

```python
import rdflib

g = rdflib.Graph()
result = g.parse('mappingbased_literals_en.ttl', format='n3')
```

```python
rows = []
with open('mappingbased_literals_en.ttl', 'r', encoding="utf8") as f:
    f.readline()
    for l in f.readlines():
        #print(l)
        try:
            split = l.split('"')
            split2 = split[0].split(' ')
            rows.append((split2[0], split2[1], split[1]))
        except:
            print(l)
        #break
        
#rows
```

```python
len(rows)
```

```python
rows[-1]
```

```python
df = pd.DataFrame.from_records(rows, columns=['from', 'relation', 'to'])
```

```python
df
```

```python
df['relation'].unique()
```

```python
import re

stats = []

for key,group in df.groupby('relation'):
    #print(key)
    #print(list(group.head(10)['to']))
    total = 0
    letters = 0
    subset = group.sample(n=100 if len(group) > 100 else len(group))['to']
    for w in subset:
        total += len(w)
        letters += len(re.findall(r'[a-zA-Z]', w))
    #print(key, letters/total)
    stats.append((key, total/len(subset), letters/total, len(group)))
```

```python
stats_df = pd.DataFrame.from_records(stats, columns=['relation', 'average_length', 'letter_percent', 'n'])

stats_df
```

```python
stats_df[(stats_df['n'] >= 10) & (stats_df['letter_percent'] > .6) & (stats_df['average_length'] > 2)]
```

```python
relations = stats_df[(stats_df['n'] >= 10) & (stats_df['letter_percent'] > .6) & (stats_df['average_length'] > 2)]['relation']
```

```python
df_subset = df[df['relation'].isin(relations)]
```

```python
df_subset['from'] = df_subset['from'].str.strip(r'[<>]').str.split('__', expand=True)[0].str.split('/', expand=True)[4].str.replace('_', ' ')
```

```python
df_subset.to_csv('dbpedia_literals_cleaning.csv', index=False)
```
