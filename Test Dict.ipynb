```python
import pickle
```

```python
G = pickle.load(open('mapping.pickle', 'rb'))
```

```python
freqs = {}
for w in G:
    for w2 in G[w]:
        if w2 not in freqs:
            freqs[w2] = 1
        else:
            freqs[w2] += 1
```

```python
freqs['north']
```

```python
P1 = 1
P2 = -3
A = -10

board = {
    'web': P1,
    'ambulance': P1,
    'hospital': P1,
    'spell': P1,
    'lock': P1,
    'charge': P1,
    'tail': P1,
    'link': P1,
    'cook': P1,
    'time': P2,
    'france': P2,
    'mexico': P2,
    'date': P2,
    'yard': P2,
    'face': P2,
    'straw': P2,
    'plate': P2,
    'food': A
}
results = {}

def add_word(w, value, ow):
    if freqs[w] < 1000:
        return
    
    if w in results:
        #results[w] += [ow]
        results[w][0] += value
        results[w][1].append((ow, G[ow][w]))
    else: 
        #results[w] = [ow]
        results[w] = [value, [(ow, G[ow][w])]]

for word in board:
    #for ass in df[df.word1 == word].word2.unique():
    for ass in G[word]:
        add_word(ass, board[word]*G[word][ass]/freqs[word], word)
```

```python
import pandas as pd
df = pd.DataFrame.from_records(list(results.items()))

def score(row):
    word = row[0]
    r = row[1]
    return sum(bool(x[1])*board[x[0]]*freqs[word] for x in r[1])
    #return sum((x[1]*board[x[0]] > 0) for x in r[1])
    
    #return sum(math.log(x[1], 10)/freqs[word]*board[x[0]] for x in r[1])
    
df['score'] = df.apply(score, axis=1)

df.sort_values('score', ascending=False)
```
