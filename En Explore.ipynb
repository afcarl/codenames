```python
import pandas as pd
import networkx as nx
```

```python
df = pd.read_csv('en.csv.gz', sep='\t')[['relation', 'to', 'from']]
df = df[(df['from'].str.extract(r'/./(.*)/.*/.*') == 'en')  & (df['to'].str.extract(r'/./(.*)/.*') == 'en')]
df['relation'] = df['relation'].str.extract(r'/./(.*)')
```

```python
df['to'] = df['to'].str.extract(r'/./.*/(.*)')
df['from'] = df['from'].str.extract(r'/./.*/(.*)/.*')
```

```python
df.relation.unique()
```

```python
df = pd.concat([
    df.rename(columns={'to': 'word1', 'from': 'word2'}),
    df.rename(columns={'to': 'word2', 'from': 'word1'})
]).drop_duplicates().reset_index(drop=True)
```

```python
df = df[['relation', 'word1', 'word2']]
```

```python
df[(df['word1'] == 'time') & (df['word2'] == 'times')]
```

```python
df = df[(df['from'].str.extract(r'/./(.*)/.*/.*') == 'en')  & (df['to'].str.extract(r'/./(.*)/.*') == 'en')]
```

```python
#board = ['time', 'france', 'mexico', 'date', 'yard', 'face', 'straw', 'plate', 'undertaker']
board = ['web', 'ambulance', 'hospital', 'spell', 'lock', 'charge', 'tail', 'link', 'cook']
board = {
    'web': 1,
    'ambulance': 1,
    'hospital': 1,
    'spell': 1,
    'lock': 1,
    'charge': 1,
    'tail': 1,
    'link': 1,
    'cook': 1,
    'time': -3,
    'france': -3,
    'mexico': -3,
    'date': -3,
    'yard': -3,
    'face': -3,
    'straw': -3,
    'plate': -3,
    'food': -10
}
results = {}

def add_word(w, value):
    if w in results:
        results[w] += value
    else: 
        results[w] = value

for word in board:
    for ass in df[df.word1 == word].word2:
        add_word(ass, board[word])
        
def validate_word(word, board):
    for b in board:
        if 'FormOf' in list(df[(df['word1'] == word) & (df['word2'] == b)]['relation']):
            return False
        
    #print(word)
    return True

for word,sim in sorted(list(results.items()), key=lambda x: -x[1])[:10]:
    if validate_word(word, board):
        print(word,sim)
```
