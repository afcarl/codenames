```python
import pandas as pd
import re
```

```python
rows = []
with open('mappingbased_objects_en.ttl', 'r', encoding="utf8") as f:
    f.readline()
    for l in f.readlines():
        try:
            split = l.split(' ')
            rows.append((
                split[0].split('/')[-1],#.replace('>', ''),
                split[1],
                split[2].split('/')[-1]#.replace('>', '')
            ))
        except:
            print(l)
            break
```

```python
df = pd.DataFrame.from_records(rows[:-1], columns=['from', 'relation', 'to'])
```

```python
df
```

```python
df['from'] = df['from'].str.strip(r'[<>]').str.split('__', expand=True)[0].str.replace('_', ' ')
```

```python
df['to'] = df['to'].str.strip(r'[<>]').str.split('__', expand=True)[0].str.replace('_', ' ')
```

```python
df
```

```python
df.to_csv('dbpedia_objects_cleaning.csv', index=False)
```
