```python
def validate_word(word, board):
    for b in board:
        if word in b or b in word:
            return False
        
    return True
```

```python
import pandas as pd

positive=['cook', 'web', 'ambulance', 'hospital', 'spell', 'lock', 'charge', 'tail', 'link']
negative=['cat', 'button', 'pipe', 'pants', 'mount', 'game', 'stick', 'file', 'worm', 'doctor']

# POSITIVE

dfs = []

for p in positive:
    dfs.append(pd.DataFrame.from_records(model.most_similar(positive=p, topn=1000), columns=['word', 'similarity']))
    
results = dfs.pop()

for df in dfs:
    
    results = results.merge(df, on='word', how='outer').fillna(0)
    results['similarity'] = results['similarity_x'] + results['similarity_y']
    results = results[['word', 'similarity']]
    
# NEGATIVE
    
dfs = []
for n in negative:
    dfs.append(pd.DataFrame.from_records(model.most_similar(positive=n, topn=1000), columns=['word', 'similarity']))
    

for df in dfs:
    results = results.merge(df, on='word', how='outer').fillna(0)
    results['similarity'] = results['similarity_x'] - 3*results['similarity_y']
    results = results[['word', 'similarity']]
    
for i,row in results.sort_values('similarity', ascending=False).head(50).iterrows():
    if validate_word(row['word'], positive+negative):
        print(row['word'], row['similarity'])
```
