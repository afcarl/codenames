```python
import gensim

# Load Google's pre-trained Word2Vec model.

model = gensim.models.KeyedVectors.load_word2vec_format('GoogleNews-vectors-negative300.bin', binary=True, limit=50000)  

#model = gensim.models.KeyedVectors.load_word2vec_format('twitter_word2vec.txt', limit=200000)  

#model = gensim.models.KeyedVectors.load_word2vec_format('glove_word2vec.txt', limit=100000)  
```

```python
model.most_similar_cosmul(
    positive=['web', 'ambulance', 'hospital', 'spell', 'lock', 'charge', 'tail', 'link', 'cook'], 
    negative=['cat', 'button', 'pipe', 'pants', 'mount', 'game', 'stick', 'file', 'worm'],
    topn=100#, restrict_vocab=100000
)
```

```python
model.most_similar(
    positive=['web', 'ambulance', 'hospital', 'spell', 'lock', 'charge', 'tail', 'link', 'cook'], 
    negative=['cat', 'button', 'pipe', 'pants', 'mount', 'game', 'stick', 'file', 'worm'],
    topn=100#, restrict_vocab=100000
)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame.from_records(model.similar_by_word('hospital', topn=1000), columns=['word', 'similarity'])
df2 = pd.DataFrame.from_records(model.similar_by_word('ambulance', topn=1000), columns=['word', 'similarity'])
df3 = pd.DataFrame.from_records(model.similar_by_word('pants', topn=1000), columns=['word', 'similarity'])
m = df.merge(df2, on='word').merge(df3, on='word')
m['avg'] = (m['similarity'] + m['similarity_x'] + m['similarity_y'])/3
m.sort_values('avg', ascending=False)
```

```python
df = pd.DataFrame.from_records(model.similar_by_word('web', topn=1000), columns=['word', 'similarity'])
df2 = pd.DataFrame.from_records(model.similar_by_word('link', topn=1000), columns=['word', 'similarity'])
m = df.merge(df2, on='word')
m['avg'] = (m['similarity_x'] + m['similarity_y'])/2
m.sort_values('avg', ascending=False)
```

```python
import itertools

list(itertools.combinations(['web', 'ambulance', 'hospital', 'spell', 'lock', 'charge', 'tail', 'link', 'cook'], 2))
```

```python
for x,y in itertools.combinations(['web', 'ambulance', 'hospital', 'spell', 'lock', 'charge', 'tail', 'link', 'cook'], 2):
    df = pd.DataFrame.from_records(model.similar_by_word(x, topn=1000), columns=['word', 'similarity'])
    df2 = pd.DataFrame.from_records(model.similar_by_word(y, topn=1000), columns=['word', 'similarity'])
    #df3 = pd.DataFrame.from_records(model.similar_by_word('pants', topn=1000), columns=['word', 'similarity'])
    m = df.merge(df2, on='word')#.merge(df3, on='word')
    #m['avg'] = (m['similarity'] + m['similarity_x'] + m['similarity_y'])/3
    m['avg'] = (m['similarity_x'] + m['similarity_y'])/2
    
    print(x,y)
    print(m.sort_values('avg', ascending=False).head(5))
    print('*'*20)
```
