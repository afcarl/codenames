```python
rows = []
with open('mappingbased_objects_en.ttl', 'r', encoding="utf8") as f:
    f.readline()
    for l in f.readlines():
        try:
            split = l.split(' ')
            rows.append((
                split[0].split('/')[-1],#.replace('>', ''),
                split[2].split('/')[-1]#.replace('>', '')
            ))
        except:
            print(l)
            break
            
df = pd.DataFrame.from_records(rows[:-1], columns=['from', 'to']).drop_duplicates()
```

```python
for col in ['from', 'to']:
    df[col] = df[col].str.strip(
        r'[<>]'
    ).str.split(
        '__', expand=True
    )[0].str.replace(
        '_', ' '
    ).str.replace(
        r'\(.*\)', ''
    ).str.replace(
        r'\s+', ' '
    ).str.replace(
        '%22', '"'
    ).str.replace(
        '%3F', '?'
    ).str.strip().str.lower().str.replace(r'[^\w ]+', '')
```

```python
text = ' '.join(df['from']) + ' ' + ' '.join(df['to'])
```

```python
import nltk
allWords = nltk.tokenize.word_tokenize(text)
allWordDist = nltk.FreqDist(allWords)
```

```python
#G=nx.Graph()
G = dict()
def add_edge(a,b):
    if a not in G:
        G[a] = {}
    if b not in G[a]:
        G[a][b] = 0
        
    G[a][b] += 1
    
def add_edges(a,b):
    add_edge(a,b)
    add_edge(b,a)
    
for index, row in df.iterrows():
    f = row['from']
    t = row['to']
    
    if f == t:
        continue
    
    #if f in dictionary and t in dictionary:
    #    add_edges(f, t)
        
    fwords = f.split(' ')
    twords = t.split(' ')
    for i in set(list(range(1, min(4, len(fwords)+1))) + [len(fwords)]):
        for j in set(list(range(1, min(4, len(twords)+1))) + [len(twords)]):
            fgrams = [y for y in [' '.join(x) for x in ngrams(fwords, i)] if y in dictionary]
            tgrams = [y for y in [' '.join(x) for x in ngrams(twords, j)] if y in dictionary]
            
            for f2 in fgrams:
                for t2 in tgrams:
                    if f2 == t2:
                        continue
                        
                    if f2 in dictionary and t2 in dictionary and f2 not in sw and t2 not in sw:
                        add_edges(f2, t2)
                        
    if index%100000 == 0:
        print(index)
```
